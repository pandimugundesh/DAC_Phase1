```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('/content/statsfinal.csv')  # Replace 'your_data.csv' with the actual file path

# Extract relevant columns
date = data['Date']
product_sales = data[['Q-P1', 'Q-P2', 'Q-P3', 'Q-P4']]

# Create a bar chart for product sales
plt.figure(figsize=(12, 6))
for product in product_sales.columns:
    plt.bar(date, product_sales[product], label=product)

plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Product Sales Analysis - Bar Chart')
plt.xticks(rotation=45)
plt.legend()

# Create a line chart for a specific product (e.g., Q-P1)
selected_product = 'Q-P1'
plt.figure(figsize=(12, 6))
plt.plot(date, product_sales[selected_product], marker='o', label=selected_product, color='green')

plt.xlabel('Date')
plt.ylabel('Sales')
plt.title(f'{selected_product} Sales Analysis - Line Chart')
plt.xticks(rotation=45)
plt.legend()

# Show both charts
plt.tight_layout()
plt.show()
```
